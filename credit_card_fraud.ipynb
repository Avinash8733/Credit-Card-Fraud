```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset to a pandas dataframe
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# first 5 rows of dataset
credit_card_data.head()
```

```python
# dataset imformation
credit_card_data.info()
```

```python
#checking number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transction & fraudulent transction
credit_card_data['Class'].value_counts()
```

This dataset is highly unblanced

0 -> Normal transaction
1 -> fraudulent transaction



```python
#seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]

```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaction

credit_card_data.groupby('Class').mean()
```

building a sample dataset containing similar distribution of normal transaction and fraudulent transaction

```python
legit_sample = legit.sample(n=492)
```

Concating two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

spliting the data into features & Targets

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

split data into training data & testing data


```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

model trsaing

```python
#logistic regression
model = LogisticRegression()
```

```python
#traing the logistic regression model with training data
model.fit(X_train,Y_train)
```

Model Evalution

```python
#accuracy score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on traing data : ',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('accuracy score on test data :',test_data_accuracy)
```
